Benchmark Notebook: P3-hull Algorithms
======================================

Use this notebook to visualize and compare the performance of
RAND-P3, MTS-M, and TIP-DECOMP-R on various graph classes.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load benchmark results
df = pd.read_csv("../benchmark_results.csv")

# Display overview
display(df.head())
```

```python
# --- Hull size vs Algorithm ---
plt.figure(figsize=(8,5))
sns.barplot(data=df, x="graph", y="hull_size", hue="algorithm")
plt.title("Hull size comparison across graph families")
plt.xlabel("Graph")
plt.ylabel("Hull set size")
plt.legend(title="Algorithm")
plt.tight_layout()
plt.show()
```

```python
# --- Runtime comparison ---
plt.figure(figsize=(8,5))
sns.barplot(data=df, x="graph", y="runtime_sec", hue="algorithm")
plt.title("Runtime comparison across algorithms")
plt.xlabel("Graph")
plt.ylabel("Runtime (seconds)")
plt.legend(title="Algorithm")
plt.tight_layout()
plt.show()
```

```python
# --- Statistical summary ---
summary = df.groupby("algorithm")[["hull_size", "runtime_sec"]].mean().round(3)
display(summary)
```
